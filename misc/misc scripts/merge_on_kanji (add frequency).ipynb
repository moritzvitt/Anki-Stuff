```python
import pandas as pd

frequency_df = pd.read_excel(
    '/Users/moritzvitt/src/Kanji Frequency List.xlsx',
)

print(frequency_df.head())
```

```python
#load Japanese Heisig__Deck in progress.txt into a df, naming it anki_df.


anki_df = pd.read_csv(
    '/Users/moritzvitt/src/Japanese Heisig__Deck in progress.txt',
    sep='\t',
    header=None,
    skiprows=4
)

print(anki_df.head())
```

```python
# Extract relevant columns from frequency_df
freq_subset = frequency_df.iloc[:, [0, 8]].copy()

# Rename columns for clarity
freq_subset.columns = ["kanji", "frequency"]

# Also rename kanji column in anki_df for clarity
anki_df = anki_df.copy()
anki_df.rename(columns={anki_df.columns[1]: "kanji"}, inplace=True)

# Merge
merged_df = pd.merge(
    anki_df,
    freq_subset,
    on="kanji",
    how="left"
)

print(merged_df.head())
```

```python
#drop the first column of the merged df, which is just an index.
merged_df.drop(merged_df.columns[0], axis=1, inplace=True)

merged_df.to_csv(
    '/Users/moritzvitt/src/Japanese Heisig__Deck in progress with frequency.csv',
    sep='\t',
    index=False,
    header=False
)
```
