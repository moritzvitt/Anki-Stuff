```python
import pandas as pd
from pathlib import Path

path = Path("~/Documents/Selected Notes.txt").expanduser()

df = pd.read_csv(
    path,
    sep="\t",
    dtype=str,
    engine="python",
    quoting=3,
    skiprows=4,
    header=None,
    names=range(36)   # force 32 columns
).fillna("")

print(df)
```

```python
import re

def make_cloze(row):
    text = row.iloc[1]     # 2nd column
    gloss = row.iloc[10]   # 11th column

    m = re.search(r"<strong>(.*?)</strong>", text)
    if not m:
        return text

    word = m.group(1)

    return re.sub(
        r"<strong>.*?</strong>",
        f"{{{{c1::{word}::{gloss}}}}}",
        text,
        count=1
    )

df["Cloze_new"] = df.apply(make_cloze, axis=1)

print(df["Cloze_new"])
```

```python
# create new dataframe with desired columns
new_df = df.iloc[:, [1]].copy()   # previous column
new_df["cloze"] = df["Cloze_new"]

df = new_df
```

```python
from pathlib import Path

out = Path("~/Documents/Selected Notes_cloze.txt").expanduser()

df.to_csv(out, sep="\t", index=False, header=False)

print("saved →", out)
```
